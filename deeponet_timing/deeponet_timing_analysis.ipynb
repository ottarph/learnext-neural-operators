```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

DATA_DIR = Path("output/data")
OUTPUT_DIR = Path("output/figures")
```

```python
timings: np.ndarray = np.load(DATA_DIR / "deeponet_timing.npy")
timings = timings[...,0,1]

dd = np.array([3, 4, 5, 6, 7], dtype=float)
ww = np.array([128, 256, 512, 1024], dtype=float)
pp = np.array([8, 16, 32, 64], dtype=float)

print(timings)
print(timings.shape)
```

```python
marg_d = timings.mean(axis=(1,))
# print(marg_d)
plt.figure()
plt.loglog(dd, marg_d, 'k-o')
a, b = np.polyfit(np.log(dd), np.log(marg_d), deg=1)
print(f"T = {np.exp(b):.1e} * x ** {a:.2f})")

marg_w = timings.mean(axis=(0,))
# print(marg_w)
plt.figure()
plt.loglog(ww, marg_w, 'k-o')
a, b = np.polyfit(np.log(ww), np.log(marg_w), deg=1)
print(f"T = {np.exp(b):.1e} * x ** {a:.2f})")

# marg_p = timings.mean(axis=(0,1))
# # print(marg_p)
# plt.figure()
# plt.loglog(pp, marg_p, 'k-o')
# a, b = np.polyfit(np.log(pp), np.log(marg_p), deg=1)
# print(f"T = {np.exp(b):.1e} * x ** {a:.2f})")
```

```python

for k in range(len(dd)):
    marg_w = timings[k,:]
    a, b = np.polyfit(np.log(ww), np.log(marg_w), deg=1)
    print(f"T = {np.exp(b):.1e} * x ** {a:.2f})")
```

```python
for k in range(len(ww)):
    marg_d = timings[:,k]
    a, b = np.polyfit(np.log(dd), np.log(marg_d), deg=1)
    print(f"T = {np.exp(b):.1e} * x ** {a:.2f})")
```
